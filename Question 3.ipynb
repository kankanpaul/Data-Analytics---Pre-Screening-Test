```python
data = pd.read_csv(r'C:\Users\paulk\Downloads\stock_prices.csv', parse_dates=['date'], dayfirst=True)

data['date'] = pd.to_datetime(data['date'])

# Calculate the rolling 30-day moving average
data['30_day_moving_avg'] = data['closing_price'].mean()
# Plot the closing price and moving average
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['closing_price'], label='Closing Price', color='blue')
plt.plot(data.index, data['30_day_moving_avg'], label='30-Day Moving Avg', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices Analysis')
plt.legend()
plt.grid(True)
plt.show()

# Time series decomposition (trend, seasonality, residual)

plt.show()
```
