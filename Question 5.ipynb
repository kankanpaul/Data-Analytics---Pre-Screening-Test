```python
import pandas as pd
```

```python
# Read the dataset
data = pd.read_csv(r'C:\Users\paulk\Downloads\customer_data.csv')
```

```python
# Handle missing values by filling NaNs in 'purchases' column with 0
data['purchases'].fillna(0, inplace=True)
```

```python
data
```

```python
# Remove duplicates based on 'customer_id'
data.drop_duplicates(subset='customer_id', keep='first', inplace=True)
```

```python
data
```

```python
# Calculate the average number of purchases for each gender
avg_purchases_by_gender = data.groupby('gender')['purchases'].mean()
```

```python
# Print the results
print(avg_purchases_by_gender)
```
